```python
import numpy as np
from scipy.linalg import solve
```

```python
def gauss(A,B,x,n):
    L=np.tril(A)
    U=A-L
    for i in range(n):
        x = np.dot(np.linalg.inv(L),B-np.dot(U,x))
        print (x)
    return x
```

```python
A = eval(input("Ingresa la matriz A: "))
#como np.array([[a11,a12,a13],[a21,a22,a23],[a31,a32,a33]])
B = eval(input("Ingresa la matriz B: ")) # como [b1,b2,b3]
x = eval(input("Ingresa the guess of x: ")) # como [x1,x2,x3]
n = eval(input("Ingresa el numero de iteraciones: "))
x = gauss(A,B,x,n)
print ("Solucion usando la sintaxis solve: \n", solve(A,B))
```
